# EDA

```python
plt.figure(figsize=(5,4))
fig = df_ben['Gender'].value_counts().plot(kind='bar', color=['Red','Orange'])
plt.xlabel("Gender Code")
plt.ylabel("Number share of patients\n")
plt.title("Distribution of BENEFICIARIES based on GENDER\n")
```

```python
df_ben['DOB'] = pd.to_datetime(df_ben['DOB'], format="%Y-%m-%d")

df_ben['Age']=df_ben['DOB'].dt.year
```

## Benificiary Age Year

```python
plt.figure(figsize=(30,20))
ax=sns.countplot('Age',data=df_ben)
plt.xlabel('Age')
plt.xticks(rotation=90)
plt.ylabel('Count')
```

## Race

It indicate majority of benificiaries 

```python
plt.figure(figsize=(6,5))
ax=sns.countplot('Race',data=df_ben)
plt.xlabel('Race')
plt.ylabel('Count')
```

```python
df_ben['month']=df_ben['DOB'].dt.month
```

```python
plt.figure(figsize=(30,20))
ax=sns.countplot('month',data=df_ben)
plt.xlabel('Age')
plt.xticks(rotation=90)
plt.ylabel('Count')
```

```python
df_ben['Dead_or_Alive']=df_ben['DOD'].apply(lambda val:0 if val !=val else 1)
```

```python
df_ben['Dead_or_Alive'].value_counts()
```

```python
plt.figure(figsize=(10,10))
ax=sns.countplot('Dead_or_Alive',data=df_ben)
plt.xlabel('Dead_or_Alive')
plt.xticks(labels=["ALIVE","DEAD"], ticks=[0,1], rotation=20)
plt.ylabel('Count')
```

```python
df_ben['DOD']=pd.to_datetime(df_ben['DOD'])
max_ben_DOD=max(df_ben['DOD'].unique()[1:])
df_ben['DOD'].fillna(value=max_ben_DOD,inplace=True)
df_ben['Age']=np.round(((df_ben['DOD']-df_ben['DOB']).dt.days)/365.0,1)
df_ben.drop(labels=['DOD'],axis=1,inplace=True)
df_ben.head()
```

```python
plt.figure(figsize=(30,20))
ax=sns.countplot('Age',data=df_ben)
plt.xlabel('Age')
plt.xticks(size=12)
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10,8))
df_ben['Age'].plot(kind='hist', color='purple')
plt.xlabel("Beneficiaries Age in years")
plt.ylabel("Frequency of patients")
plt.title("Distribution of BENEFICIARIES AGE")
```

```python
Age_condition=[
               (df_ben['Age']<=34),
                (df_ben['Age']>34) & (df_ben['Age']<=44),
               (df_ben['Age']>=45)&(df_ben['Age']<=64),
              (df_ben['Age']>=65)]

column_values=['Young','Mid','Old','Very Old']
df_ben['Age']=np.select(Age_condition,column_values)

plt.figure(figsize=(7,5))
ax=sns.countplot('Age',data=df_ben)
plt.xlabel('Age')
plt.ylabel('Count')
```
